Supervised Scaling

```python
#target
data['SnowDay'] = (data['BASEL_snow_depth'] > 0).astype(int)
```

```python
numeric_columns1 = data.select_dtypes(include=['float64', 'int64']).columns.tolist()
columns_to_exclude1 = ['DATE', 'MONTH', 'SnowDay']  # Now including the target
features_to_scale1 = [col for col in numeric_columns1 if col not in columns_to_exclude1]
```

```python
# Scale the features separately for snow and no-snow days
scaled_snowday = []

for snow_flag in data['SnowDay'].unique():
    subset = data[data['SnowDay'] == snow_flag]
    
    # Scale only the selected features
    scaler1 = StandardScaler()
    scaled_features1 = scaler1.fit_transform(subset[features_to_scale1])
    
    # Convert back to DataFrame
    scaled_df2 = pd.DataFrame(scaled_features1, columns=features_to_scale1)
    
    # Add back the SnowDay flag and any columns you excluded
    for col in columns_to_exclude1:
        if col in subset.columns:
            scaled_df2[col] = subset[col].values
    
    scaled_snowday.append(scaled_df2)

# Merge groups together
supervised_scaled_snowdays = pd.concat(scaled_snowday).reset_index(drop=True)
```
